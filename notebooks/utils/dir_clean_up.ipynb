```python
import shutil
from pathlib import Path

# Project root
PROJECT_ROOT = Path(r"C:\Users\chira\Desktop\data_engineering\PySpark\nyc-taxi-analytics-platform")

# Paths to clean
paths_to_clean = [
    PROJECT_ROOT / "data" / "bronze" / "_processed_files",
    PROJECT_ROOT / "data" / "bronze" / "nyc_taxi" / "green"
]

print("="*60)
print("CLEANUP: Removing corrupted bronze data")
print("="*60)

for path_to_clean in paths_to_clean:
    if path_to_clean.exists():
        try:
            # Force remove entire directory tree
            shutil.rmtree(path_to_clean, ignore_errors=True)
            print(f"✓ Removed: {path_to_clean}")
        except Exception as e:
            print(f"✗ Failed to remove {path_to_clean}: {e}")
            # Try alternative method
            try:
                import os
                os.system(f'rmdir /S /Q "{path_to_clean}"')
                print(f"  ✓ Removed using system command")
            except Exception as e2:
                print(f"  ✗ System command also failed: {e2}")
    else:
        print(f"⚠ Path doesn't exist: {path_to_clean}")

print("\n" + "="*60)
print("Cleanup complete!")
print("="*60)
```
